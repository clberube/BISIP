```python
from bisip import mcmcinv
```

```python
# First test run
# To run a black-box type of stochastic inversion you need two things:

""" 1.
    Model to use ?"""
model = "ColeCole" # Cole-Cole (double Cole-Cole by default)

""" 2.
    Paths to files ?"""
filepath = "/Users/Charles/Documents/SIP dat files/AVG_SIP-Reciprocals_K389737.dat"
    
```

```python
# This will run the stochastic inversion on the file using all default MCMC parameters
# This shouldn't take long. For now...
sol = mcmcinv(model, filepath)
```

```python
sol.plot_fit() # Use this to plot the fit results
```

```python
# That's pretty good. We only ran 10 000 iterations. 
# The 95% high probability density interval shows that the model
# is more uncertain in the frequency range where the data is noisy. Cool!
# The results could probably be better if we ran more iterations, or if we used less noisy data.
# Let's see how in Example 2.
```
